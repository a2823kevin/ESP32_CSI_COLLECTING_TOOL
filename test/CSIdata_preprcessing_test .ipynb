```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# default fs=100
# jump_all fs=76.67535654040796
# nomove_all fs=144.3556924261379
# walk_all fs=72.35890014471774
```

```python
df_jp = pd.read_csv('jump_all.csv')
display(df_jp)
```

```python
# 實虛合併
sub = 1
df1 = pd.DataFrame({'0' : []})
n = np.zeros((15000,1), dtype=complex)
for i in range(0, 128, 2):
    df = pd.DataFrame(n, columns=["subcarrier"+str(sub)])
    df["subcarrier"+str(sub)] =  df_jp[str(i+1)] + (1j*df_jp[str(i)])
    df1 = pd.concat([df1, df], axis=1)
    sub+=1
df2 = df1.drop(["0"], axis=1)
df2
```

```python
# inverse fft to time domain
t = np.fft.ifft(df2)
column_name = []
for i in range(64):
    column_name.append("subcarrier"+str(i+1))
print(column_name)
# df3:time domain
df3 = pd.DataFrame(t.real, columns=column_name)
df3
```

# Butterworth low pass filter

```python
from scipy.signal import butter, lfilter, freqz

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
order = 8
fs = 100
cutoff = 30
```

```python
b, a = butter_lowpass(cutoff, fs, order)
w, h = freqz(b, a, worN=8000)
plt.subplot(2, 1, 1)
plt.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
plt.plot(cutoff, 0.5*np.sqrt(2), 'ko')
plt.axvline(cutoff, color='k')
plt.xlim(0, 0.5*fs)
plt.title("Lowpass Filter Frequency Response")
plt.xlabel('Frequency [Hz]')
plt.grid()
```

```python
b, a = butter_lowpass(cutoff, fs, order)
y = butter_lowpass_filter(df3, cutoff, fs, order)
```

```python
dfy = pd.DataFrame(y)
display(dfy)
```
